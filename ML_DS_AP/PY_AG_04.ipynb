## Merge, Concat, Join

Os objetos Series e DataFrame em pandas são ferramentas poderosas para explorar e analisar dados. Parte de seu poder vem de uma abordagem multifacetada para combinar conjuntos de dados separados. Com os pandas, você pode mesclar, juntar e concatenar seus conjuntos de dados, permitindo unificar e entender melhor seus dados enquanto os analisa.

- merge() para combinar dados em colunas ou índices comuns
- join() para combinar dados em uma coluna chave ou um índice
- concat () para combinar DataFrames em linhas ou colunas

Você pode usar merge() sempre que quiser uma funcionalidade semelhante às operações de junção de um banco de dados. É a mais flexível das três operações que você aprenderá.

Quando você deseja combinar objetos de dados com base em uma ou mais chaves, semelhante ao que faria em um banco de dados relacional, merge() é a ferramenta de que você precisa. Mais especificamente, merge() é mais útil quando você deseja combinar linhas que compartilham dados.

Você pode obter junções muitos-para-um e muitos-para-muitos com merge(). Em uma junção muitos-para-um, um de seus conjuntos de dados terá muitas linhas na coluna de mesclagem que repetem os mesmos valores. Por exemplo, os valores podem ser 1, 1, 3, 5 e 5. Ao mesmo tempo, a coluna de mesclagem no outro conjunto de dados não terá valores repetidos. Tome 1, 3 e 5 como exemplo.

Como você deve ter adivinhado, em uma junção muitos-para-muitos, ambas as colunas de mesclagem terão valores repetidos. Essas mesclagens são mais complexas e resultam no produto cartesiano das linhas unidas.

Isso significa que, após a mesclagem, você terá todas as combinações de linhas que compartilham o mesmo valor na coluna-chave. Você verá isso em ação nos exemplos abaixo.

O que torna merge() tão flexível é o grande número de opções para definir o comportamento de sua mesclagem. Embora a lista possa parecer assustadora, com a prática você será capaz de mesclar habilmente conjuntos de dados de todos os tipos.

Ao usar merge(), você fornecerá dois argumentos obrigatórios:

 - left DataFrame
 - right DataFrame

Depois disso, você pode fornecer vários argumentos opcionais para definir como seus conjuntos de dados são mesclados:

como define que tipo de fusão fazer. O padrão é  'inner', mas outras opções possíveis incluem 'outer', 'left', and 'right'.

- on: informa merge() em quais colunas ou índices, também chamados de colunas-chave ou índices-chave, você deseja unir. Isso é opcional. Se não for especificado, e left_index e right_index (abordados abaixo) forem False, as colunas dos dois DataFrames que compartilham nomes serão usadas como chaves de junção. Se você usar on, a coluna ou índice especificado deverá estar presente em ambos os objetos.

- left_on e right_on: especificam uma coluna ou índice que está presente apenas no objeto esquerdo ou direito que você está mesclando. O padrão de ambos é Nenhum.

- left_index e right_index: são ambos padronizados como False, mas se você quiser usar o índice do objeto esquerdo ou direito a ser mesclado, poderá definir o argumento relevante como True.

- suffixes: é uma tupla de strings para anexar a nomes de colunas idênticos que não são chaves de mesclagem. Isso permite que você acompanhe as origens das colunas com o mesmo nome.

```
pd.merge(tabela,tabela2, on=['Class'])
```

```
pd.merge(tabela,tabela2, on=['Class','B2'])
```

### Inner, Left, Right, Outer

left_right.png

Neste exemplo, você usará merge() com o argumento inner, o que resultará em uma junção interna. Lembre-se de que em uma junção interna, você perderá linhas que não correspondem à coluna de chave do outro DataFrame.

```
pd.merge(tabela,tabela2, on='Class', how='inner')
```

Neste exemplo, você especificará uma junção left, também conhecida como junção externa esquerda, com o parâmetro how. O uso de uma junção externa esquerda deixará seu novo DataFrame mesclado com todas as linhas do DataFrame esquerdo, enquanto descarta as linhas do DataFrame direito que não têm uma correspondência na coluna de chave do DataFrame esquerdo.

```
pd.merge(tabela,tabela2, on='Class', how='left')
```

A junção right, ou junção externa direita, é a versão espelhada da junção esquerda. Com esta junção, todas as linhas do DataFrame direito serão retidas, enquanto as linhas no DataFrame esquerdo sem uma correspondência na coluna de chave do DataFrame direito serão descartadas.

```
pd.merge(tabela,tabela2, on='Class', how='right')
```

Aqui, você especificará um outer no parâmetro how. Lembre-se dos diagramas acima que em uma junção externa - também conhecida como junção externa completa - todas as linhas de ambos os DataFrames estarão presentes no novo DataFrame.

```
pd.merge(tabela,tabela2, on='Class', how='outer')
```

```
pd.merge(tabela, tabela2, left_on='Class', right_index=True, how='outer')
```

### Concat

A concatenação é um pouco diferente das técnicas de mesclagem que você viu acima. Com a mesclagem, você pode esperar que o conjunto de dados resultante tenha linhas dos conjuntos de dados pai misturados, geralmente com base em alguns pontos em comum. Dependendo do tipo de mesclagem, você também pode perder linhas que não possuem correspondências no outro conjunto de dados.

Com a concatenação, seus conjuntos de dados são apenas costurados ao longo de um eixo — o eixo da linha ou o eixo da coluna. Visualmente, uma concatenação sem parâmetros ao longo das linhas ficaria assim:

## Corte e categorização

A função cut() do Pandas é usada para separar os elementos da matriz em diferentes compartimentos. A função de corte é usada principalmente para realizar análises estatísticas em dados escalares.

```
NDVI = [0.5, 0.2, 0.1, 0.99, -0.5, -0.4, 0.87, 0.79, 0.24, 0.44, 0.33, 0.12, 0.83]
```

```
bins = [-1, -0.5, 0 , 0.2, 0.4, 0.6, 0.8,1]
```

```
categ = pd.cut(NDVI, bins)
```

```
categ
```

```
pd.value_counts(categ)
```

Neste exemplo definimos 7 intervalos de valores NDVI:

```
pd.cut(NDVI, bins = bins, labels=['1°intervalo', '2°intervalo', '3°intervalo', '4°intervalo', '5°intervalo', '6°intervalo','7°intervalo'])
```

Em vez de passarmos os valores específicos como Bins, podemos indicar a quantidade de divisões:

```
pd.cut(NDVI, bins=6, labels=False)
```

```
pd.cut(NDVI, bins=6, labels=False, retbins=True)
```

```
catg = pd.cut(NDVI, bins=4, labels=False, retbins=True)
```

```
catg
```

```
catg[0]
```

Além disso, no caso de variáveis categóricas, a ordem lógica não é a mesma dos dados categóricos, por exemplo "um dois três". Mas a classificação dessas variáveis usa ordem lógica.

## Dummies

No momento de realizar o treinamento de modelos de machine learning, nos deparamos com um grande problema ao lidar com variáveis categóricas. As categorias não podem ser utilizadas em forma de texto, uma vez que os algoritmos compreendem apenas valores numéricos. Também não podemos simplesmente atribuir um valor numérico para cada categoria, uma vez que esse processo, conhecido como ponderação arbitrária, pode criar uma ordenação e pesos para as categorias que não refletem a realidade.

O processo correto de transformação das variáveis categóricas é feito a partir da criação de novas colunas a partir das categorias. Cada uma delas se torna uma nova coluna e o valor na linha correspondente será 1, caso tenha a presença da característica. Do contrário, será 0. Esse processo é conhecido como codificação "one-hot".

image.png

O método get_dummies() é simples de utilizar e faz a transformação de forma direta das variáveis categóricas.

```
df = pd.DataFrame({'classe': ['soil', 'water', 'soil', 'urban', 'water', 'soil','Forest'],
                'data1': range(7)})
```

```
df
```

```
pd.get_dummies(df['classe'])
```

### GroupBy

As 3 etapas de um processo Groupby

Qualquer processo groupby envolve alguma combinação das 3 etapas a seguir:

- Dividindo o objeto original em grupos com base nos critérios definidos.
- Aplicando uma função a cada grupo.
- Combinando os resultados.

groupby.png

Vamos criar um Dataframe de exemplo:

```
df = pd.DataFrame({'classe': ['soil', 'water', 'soil', 'urban', 'water', 'soil', 'water', 'vegetation', 'vegetation', 'soil', 'vegetation','urban','water','soil','urban','vegetation'],
                    'ano': [2018, 2018, 2018, 2018, 2018, 2019, 2019, 2019, 2019, 2019, 2020, 2020, 2020, 2020, 2020, 2020],
                   'NDVI': [0.36,0.55,0.99,0.12,0.22,0.68,0.66,0.57,0.99,0.10,0.33,0.55,0.32,0.86,0.71,0.03],
                   'SAVI': [0.24,0.15,0.92,0.77,0.12,0.77,0.10,0.07,0.90,0.50,0.65,0.25,0.82,0.86,0.58,0.33]
})
```

```
df
```

Podemos calcular a média da coluna NDVI:

```
df['NDVI'].mean()
```

Aplicamos o groupby na coluna NDVI para agruparmos por ano e calculamos a média.

```
df['NDVI'].groupby(df['ano']).mean()
```

Também é possível agrupar por mais de uma coluna.

```
df.groupby(['ano','classe']).mean()
```

Agora vamos obter algumas estatísticas de uma coluna do agrupamento:

```
result = df.groupby('classe').agg({'NDVI': ['mean', 'min', 'max']})
```

```
result
```

Também é possível obter as estatísticas de várias colunas do agrupamento:

```
result = df.groupby('classe').agg({'NDVI': ['mean', 'min', 'max'], 'SAVI': ['mean', 'min', 'max']})
```

```
result
```

Para reduzir as colunas em um nível apenas, usamos isso:

```
result = result.set_axis(result.columns.map('_'.join), axis=1)
```

```
result
```
